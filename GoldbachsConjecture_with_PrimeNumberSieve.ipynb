<h1>Prime Number Sieve and Godlbach's Conjecture</h1>

Using the Prime number sieve to load out the composite numbers and then using Goldbach's conjecture, finding pairs of prime numbers that sum up to even number above 2

```python
numberPrime = dict()
```

```python
def isPrime(n):
    global numberPrime
    if n in numberPrime:
        return numberPrime[n]
    for i in range(2,n):
        if n % i == 0:
            return False
    return True
```

```python
print(isPrime(7), isPrime(4), isPrime(2), isPrime(11))
```

```python
def sieve(n):
    global numberPrime
    for i in range(2, n + 1):
        if i not in numberPrime and isPrime(i):
            numberPrime[i] = True
            for j in range(2, n//i + 1):
                if i * j <= n:
                    numberPrime[i*j] = False
                else:
                    break
```

```python
def goldbachUpto(n):
    for i in range(4, n+1, 2):
        found = False
        for p1 in range(n, 1, -1):
            if isPrime(p1):
                for p2 in range(2, n+1):
                    if isPrime(p2) and p1 + p2 == i:
                        found = True
                        print(i, "= ", p1, "+ ", p2)
                        break
            if found:
                break
```

```python
upperBound = int(input("Please enter upper bound to do Golbach's on: "))
sieve(upperBound)
goldbachUpto(upperBound)
```
